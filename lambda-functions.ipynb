Lambda functions are small functions usually not more than a line. It can have any number of arguments just like a normal function. The body of lambda functions is very small and consists of only one expression. The result of the expression is the value when the lambda is applied to an argument. Also there is no need for any return statement in lambda function.

```python
# Motivation Example
#Goal - Create function to add 3 input numerics & emit out the output.

#User Input
x=int(input("Input for X:"))
y=int(input("Input for Y:"))
z=int(input("Input for Z:"))

add_three_nos = lambda x,y,z : x+y+z #Lambda Function to create sum of 3 input numers
squr = lambda x:x**2
defultVal = lambda x,y=2: x**y

print("Sum of 3 numbers:",add_three_nos(x,y,z)) #Emit the output
print(f'squaring of {y} : {squr(y)}')
print(f'default *ing of {z} : {defultVal(z)}')
```

```python
# Motivation Example -2
#Goal : Sort a list of tuples with lambda expression

data = [("Apples", "Jammu-Kashmir", "Himachal Pradesh","Uttarakhand"), ("Pears", "Jammu-Kashmir", "Uttarakhand","Punjab"), \
        ("Oranges", "Punjab","Madhya Pradesh","Andhra Pradesh")]
data.sort(key=lambda x:x[2],reverse=True)
print(data)
```

Comparison between Standard Functions vs Lambda Functions

Standard Funtion               Lambda Function

Reusable                        Single use
Multiple lines of code          Single line of code
Named                           Named/Anonymous
None/more inputs                None/more inputs
None/more returns               One/more returns

```python
# Lambda with sort/filter/map/reduce etc

#Sort : Definition : list.sort(key=None,reverse=False)
data = [("Apples", "Jammu-Kashmir", "Himachal Pradesh","Uttarakhand"), ("Pears", "Jammu-Kashmir", "Uttarakhand","Punjab"),
        ("Oranges", "Punjab","Madhya Pradesh","Andhra Pradesh")]
data.sort(key=lambda x:x[2],reverse=True)
print(data)

#Sort - 2nd example
names = ['Sam Walton','Isaac Newton','Dhirubhai Ambani','Abdul Kalam']
names.sort(key=lambda x:x.split()[-1].lower())
print(names)
```

```python
#Sort - in Dict using Sorted

lst_dict = [{"name":"ABC","age" : 10,"city":"NYC"},
           {"name":"XYZ","age" : 20,"city":"BLR" },
            {"name":"DEF","age" : 20,"city":"CHN"}]

sort_nm = sorted(lst_dict, key=lambda x: (x["age"],x['city']), reverse=True)

print(sort_nm)
```

```python
#Sort - Dict of Dict using Lambda
key_value ={1:{"name":"ABC","age" : 10,"city":"NYC"},
           2:{"name":"XYZ","age" : 20,"city":"BLR" },
            3:{"name":"DEF","age" : 20,"city":"CHN"}}     

dct_srted = sorted(key_value.items(),key=lambda x: x[1]['city'], reverse=True)
print(dct_srted)
```

```python
#Using OrderedDict
from collections import OrderedDict

#Sort - Dict of Dict using Lambda
key_value ={1:{"name":"ABC","age" : 10,"city":"NYC"},
           2:{"name":"XYZ","age" : 20,"city":"BLR" },
            3:{"name":"DEF","age" : 20,"city":"CHN"}}     

dct_srted = OrderedDict(sorted(key_value.items(),key=lambda x: x[1]['city'], reverse=False))
print(dct_srted)
```

```python
#Sort - List of lists
mylist = [["quux", 1, "a"], ["bar", 1, "b"],["quux", 0, "c"]]
mylist.sort(key=lambda x:(x[1],x[2]),reverse=True)
print(mylist)
```

```python
#Sorting a list of objects
class Person:
    
    def __init__(self,name,age):
        self.name =name
        self.age =age
    
    def __repr__(self):
        return f'{self.name} : {self.age}'

alex = Person('Alex',12)
muthu = Person('Muthu',15)

lst_p = [alex,muthu]

print(lst_p)
lst_p.sort(key=lambda x:x.age,reverse=True)
print(lst_p)
```

```python
#Filter : filter(function,iterable)

nums = range(1,30)

odd = list(filter(lambda x: x%2 !=0,nums))

print(odd)
```

```python
#Filter
from statistics import mean

lst = list(range(1,30))
avg = mean(lst)

abv_avgLst = list(filter(lambda x : x>avg,lst))
blw_avgLst = list(filter(lambda x : x<avg,lst))

print(abv_avgLst)
print(avg)
print(blw_avgLst)
```

```python
#Filter
fullLst = ['apple','banana','orange']
chk = ['apple','fig','orange']

chkList = list(filter(lambda x : x in chk,fullLst))
lenList = list(filter(lambda x : len(x) >5,fullLst))

print(chkList)
print(lenList)
```

```python
#Map : map(function,iterable)
nums = [1,2,3,4]

sqrLst = list(map(lambda x : x**2,nums))

print(sqrLst)
```

```python
#Map to test
nums = list(range(1,10))

even = list(map(lambda x : x%2==0,nums))
print(even)
```

```python
#Map work with tuples
numbs = [('Brazil',17),('Germany',21),('France',17)]

newIncList = sorted(list(map(lambda x : (x[0],x[1]+1),numbs)),key=lambda x : (-x[1],x[0]))

print(newIncList)
```

```python
#Reduce : reduce(function,iterable)

from functools import reduce
lst = list(range(1,5))
cum_ttl = reduce(lambda x,y: x+y,lst)
print(cum_ttl)
```

```python
from functools import reduce
names = ['dan','derren','webber']
conct_nm = reduce(lambda x,y: x+'-'+y[:2],names,'')
print(conct_nm)
```

```python
#Testing Lambda Functions

import unittest
sqrd = lambda x:x**2

class lambdaTest(unittest.TestCase):
    
    def test_sqrd(self):
        self.assertEqual(sqrd(2),4)
    
    def test_zero(self):
        self.assertEqual(sqrd(0),0)
    
    def test_ng(self):
        self.assertEqual(sqrd(-2),4)

if __name__ == '__main__':
    unittest.main(argv=['first-arg-is-ignored'], exit=False,verbosity=2)  
    
```
