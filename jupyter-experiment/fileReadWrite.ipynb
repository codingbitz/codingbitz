# Read , Write & append to the file

```python
def read_write(readFile,writeFile):
  with open('/content/drive/My Drive/MyLearnings/Python/colabNotebooks/'+readFile+".txt",'r') as readFile,open('/content/drive/My Drive/MyLearnings/Python/colabNotebooks/'+writeFile+".txt",'w')  as writeFile:
    src_1 = readFile.readlines()
    for line in src_1:
      writeFile.write(line.upper())
    readFile.close()
    writeFile.close()
read_write("randomWrite","randomWrite_1") 
#Append
import datetime
tday =datetime.date.today()
writeFile='randomWrite_1'
with open('/content/drive/My Drive/MyLearnings/Python/colabNotebooks/'+writeFile+".txt",'a')  as writeFile:
  cnty="contents for: "+str(tday)+" written to file\n"
  writeFile.writelines(cnty)
  writeFile.close()
```

# Reading from CSV
DELIMITER,ESCAPECHAR,QUOTECHAR

```python
import csv
with open('/content/drive/My Drive/MyLearnings/Python/colabNotebooks/'+'animals'+".csv",'r') as open_csv:
    src_1 = csv.reader(open_csv,delimiter=',')
    line_cnt=0
    for row in src_1:
      if line_cnt ==0:
        print(f'columnName are {",".join(row)}')
        line_cnt +=1
      else:
        print(f'in rowID: {row[0]} animal present is {row[1]}')
print(f'processed {line_cnt} lines.')
```

```python
with open('/content/drive/My Drive/MyLearnings/Python/colabNotebooks/'+'animals'+".csv",'r') as open_csv:
    src_1 = csv.DictReader(open_csv,delimiter=',',quotechar='"')
    line_cnt=0
    for row in src_1:
      if line_cnt ==0:
        print(f'columnName are {",".join(row)}')
        line_cnt +=1
      print(f'in row: {row["animal_id"]} animal present is {row["name"]}')
print(f'processed {line_cnt} lines.')
```
