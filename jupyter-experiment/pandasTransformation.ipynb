```python
import os
import pathlib
base_path = str(pathlib.Path().absolute())
input_path=base_path +'\\input\\'

#List down the files
for dirname, _, filenames in os.walk(input_path):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#####- Loading Commercial loan data CSV file into pandas df & reading sample data
commercial_loan_data = pd.read_csv(input_path+"commercial_loan_data.csv")
commercial_loan_data.head(2)
commercial_loan_data.info()
```

```python
#####- Loading loan payment data CSV file into pandas df & reading sample data
loan_payment_data = pd.read_csv(input_path+"loan_payments_data.csv")
loan_payment_data.head(2) #print 1st 2 elements of df
loan_payment_data.info() # lists dataset details like no of rows/column names & its dtypes etc
loan_payment_data.shape # gives no of rows/columns for the dataset
loan_payment_data.describe() #basic summary stats for the dataset   
```

```python
#####- DF - 3 major components : values/columns/index
print(loan_payment_data.values) # values in df
print(loan_payment_data.columns) # columns in df
print(loan_payment_data.index)
```

```python
#####- Sorting & Subsetting of df

#####- Sorting by single value
loan_payment_data.sort_values("past_due_days",ascending=False).head(5) #Sort based on column values.
######- Sorting by multiple values
loan_payment_data.sort_values(["past_due_days","education"],ascending=[False,True]).head(5) #Sort based on column values.
```

```python
#####- Subsetting
loan_payment_data["education"].unique() #distinct values for attribute education.
#####- Subsetting multiple columns
lpd =loan_payment_data.drop_duplicates(['education','Gender']) #goal is to subset by multiple columns but remove dups.
lpd[['education','Gender']]
```

```python
#####- Subset rows
#Option 1:
lpd_deg_hold = loan_payment_data[(loan_payment_data["education"]=='college') | (loan_payment_data["education"]=='Bechalor')]
print(lpd_deg_hold.shape) #287/500 loan overdue are ppl with degreee/college

#Option 2
loan_payment_data[loan_payment_data["education"].isin(['college','Bechalor'])].shape
```

```python
#Adding New columns
loan_payment_data['loan_status_lower'] = loan_payment_data['loan_status'].str.lower()
print(loan_payment_data.head(2))
```
