# Small Case Study - Regex

```python
import re
import requests
the_idiot_url = 'https://www.gutenberg.org/files/2638/2638-0.txt'

def get_book(url):
    # Sends a http request to get the text from project Gutenberg
    raw = requests.get(url).text
    # Discards the metadata from the beginning of the book
    start = re.search(r"\*\*\* START OF THIS PROJECT GUTENBERG EBOOK .* \*\*\*",raw ).end()
    # Discards the text starting Part 2 of the book
    stop = re.search(r"II", raw).start()
    # Keeps the relevant text
    text = raw[start:stop]
    return text
```

```python
def preprocess(sentence):
    return re.sub('[^A-Za-z0-9.]+' , ' ', sentence).lower()

book = get_book(the_idiot_url)
processed_book = preprocess(book)
print(processed_book)
```

```python
len(re.findall(r'the', processed_book))
```

```python
processed_book = re.sub(r'\si\s', " I ", processed_book)
#print(processed_book)
```

```python
op_sp = open(r"forregex.txt", "r").read()
```

```python
# Read the file content

for line in op_sp.split("n"):
    if "From:" in line:
        print(line)
```

```python
#Need for re
import re

for line in re.findall("From:.*", op_sp):
    print(line)
```

```python
# DOT
for line in re.findall("From:.", op_sp):
    print(line)
```

```python
#DOT - no of dots
for line in re.findall("From:...........", op_sp):
    print(line)
```

```python
#DOT - no of dots
for line in re.findall("From:.*", op_sp): #greedy search
    print(line)
```

```python
# Return only the name
match = re.findall("From:.*", op_sp)

for line in match:
    print(re.findall('\".*\"', line))
```

```python
# Return email address
for line in match:
    print(re.findall("\w\S*@.*\w", line))
```

```python
recipient = re.search(r"To:.*", item)
if recipient is not None:
    r_email = re.search(r"wS*@.*w", recipient.group())
    r_name = re.search(r":.*<", recipient.group())
else:
    r_email = None
    r_name = None
```

```python
if r_email is not None:
    recipient_email = r_email.group()
else:
    recipient_email = None
    emails_dict["recipient_email"] = recipient_email

if r_name is not None:
    recipient_name = re.sub("s*<", "", re.sub(":s*", "", r_name.group()))
else:
    recipient_name = None
emails_dict["recipient_name"] = recipient_name
```
