```python
import sqlite3
```

```python
conn    = sqlite3.connect("tasky.db")
print("Opened database successfully")
curA        = conn.cursor()
```

```python
conn.execute('''CREATE TABLE blearn (learn_id int AUTO_INCREMENT PRIMARY KEY,learn_date date, learn_start_time datetime,learn_end_time datetime,area_of_study VARCHAR(255),sub_area VARCHAR(255),completion_status VARCHAR(255),learning_src TEXT,brief_desc TEXT,comments TEXT,code_practice_link TEXT,audit_ts DATETIME DEFAULT (STRFTIME('%d-%m-%Y   %H:%M', 'NOW','localtime')))''')
print("Table created successfully")
#conn.close()
```

```python
#show tables
tables =curA.execute("select name from sqlite_master where type = 'table'")
for t in tables:
    print(t)
```

```python
tasky = curA.execute("select * from blearn order by learn_id desc")
for t in tasky:
    print(t)
```
